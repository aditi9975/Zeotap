```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
transaction_df = pd.read_csv('Transactions.csv')
product_df = pd.read_csv('Products.csv')
customer_df = pd.read_csv('Customers.csv')

```

```python
print("Transaction Dataset Preview:")
print(transaction_df.head())

print("\nProduct Dataset Preview:")
print(product_df.head())

print("\nCustomer Dataset Preview:")
print(customer_df.head())
```

```python
print("\nTransaction Dataset Info:")
transaction_df.info()

print("\nProduct Dataset Info:")
product_df.info()

print("\nCustomer Dataset Info:")
customer_df.info()
```

```python
print("\nMissing Values in Transaction Dataset:")
print(transaction_df.isnull().sum())

print("\nMissing Values in Product Dataset:")
print(product_df.isnull().sum())

print("\nMissing Values in Customer Dataset:")
print(customer_df.isnull().sum())
```

```python
print("\nSummary Statistics for Transaction Dataset:")
print(transaction_df.describe())

print("\nSummary Statistics for Product Dataset:")
print(product_df.describe())

print("\nSummary Statistics for Customer Dataset:")
print(customer_df.describe())
```

```python
# Distribution of numerical variables in Transaction Dataset
numerical_cols = transaction_df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(transaction_df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col} in Transaction Dataset')
    plt.show()
```

```python
# Correlation matrix for Transaction Dataset
plt.figure(figsize=(12, 6))
numeric_cols = transaction_df.select_dtypes(include=['float64', 'int64']).columns  # Select numeric columns
correlation = transaction_df[numeric_cols].corr()  # Compute correlation on numeric columns
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Transaction Dataset')
plt.show()
```

```python
categorical_cols_product = product_df.select_dtypes(include=['object']).columns
for col in categorical_cols_product:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=product_df[col], order=product_df[col].value_counts().index)
    plt.title(f'Count of {col} in Product Dataset')
    plt.show()

categorical_cols_customer = customer_df.select_dtypes(include=['object']).columns
for col in categorical_cols_customer:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=customer_df[col], order=customer_df[col].value_counts().index)
    plt.title(f'Count of {col} in Customer Dataset')
    plt.show()
```
